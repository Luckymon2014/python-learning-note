```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 100)
y1 = 2*x + 1
y2 = x**2

plt.xlim((-1,2))
plt.ylim((-2,3))

plt.xlabel('x')
plt.ylabel('y')

plt.xticks(np.linspace(-2,2,11))
plt.yticks([-1,0,1,2,3],
          ['level1','level2','level3','level4','level5'])

l1, = plt.plot(x, y1, color='red', linewidth=1, linestyle='--')
l2, = plt.plot(x, y2, color='blue', linewidth=2.5, linestyle='-')

plt.legend(handles=[l1,l2], labels=['line1','line2'],loc='best')

plt.show()
```
