```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1, 1, 100)
y1 = 2*x + 1
y2 = x**2

plt.figure()
plt.plot(x, y1)

plt.figure()
plt.plot(x, y2)

plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.plot(x, y2)
plt.show()
```

```python
plt.figure()
plt.plot(x, y1, color='red', linewidth=1, linestyle='--')
plt.plot(x, y2, color='blue', linewidth=2.5, linestyle='-')
plt.show()
```
