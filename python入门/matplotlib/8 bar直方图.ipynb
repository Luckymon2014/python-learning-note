```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x,y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x,-y)
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x,y,facecolor='#eeeeee',edgecolor='white')
plt.show()
```

```python
x = np.arange(10)
y = 2**x + 10

plt.bar(x,y)

for x,y in zip(x,y):
    plt.text(x,y+0.4,'%.2f' % y, ha='center', va='bottom') #+0.4调整位置

plt.show()
```
